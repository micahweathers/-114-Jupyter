## Stack Implementation

```python
class Stack:
    def __init__(self):
        self.items = []
    
    def push(self, item):
        self.items.append(item)
    
    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        return None
    
    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        return None
    
    def is_empty(self):
        return len(self.items) == 0
    
    def size(self):
        return len(self.items)
```

```python
# Test Stack
stack = Stack()
print(f"Empty: {stack.is_empty()}")

stack.push(1)
stack.push(2)
stack.push(3)
print(f"Size: {stack.size()}")
print(f"Peek: {stack.peek()}")
print(f"Pop: {stack.pop()}")
print(f"Size: {stack.size()}")
```

## Queue Implementation

```python
class Queue:
    def __init__(self):
        self.items = []
    
    def enqueue(self, item):
        self.items.append(item)
    
    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)
        return None
    
    def peek(self):
        if not self.is_empty():
            return self.items[0]
        return None
    
    def is_empty(self):
        return len(self.items) == 0
    
    def size(self):
        return len(self.items)
```

```python
# Test Queue
queue = Queue()
print(f"Empty: {queue.is_empty()}")

queue.enqueue('A')
queue.enqueue('B')
queue.enqueue('C')
print(f"Size: {queue.size()}")
print(f"Peek: {queue.peek()}")
print(f"Dequeue: {queue.dequeue()}")
print(f"Size: {queue.size()}")
```

## Hangman Game

```python
import random

def choose_word():
    words = ['apple', 'banana', 'orange', 'grape', 'pear']
    return random.choice(words)

def show_word(secret_word, guessed_letters):
    display_word = ''
    for letter in secret_word:
        if letter in guessed_letters:
            display_word += letter + ' '
        else:
            display_word += '_ '
    return display_word.strip()

def play_hangman():
    secret_word = choose_word()
    guessed_letters = set()
    incorrect_guesses = 0
    max_incorrect = 6
    
    print("Welcome to Hangman!")
    print(f"The word has {len(secret_word)} letters.")
    print(f"You have {max_incorrect} incorrect guesses allowed.\n")
    
    while incorrect_guesses < max_incorrect:
        current_display = show_word(secret_word, guessed_letters)
        print(f"Word: {current_display}")
        print(f"Guessed: {sorted(guessed_letters)}")
        print(f"Remaining: {max_incorrect - incorrect_guesses}\n")
        
        if '_' not in current_display:
            print(f"You won! The word was: {secret_word}")
            return
        
        guess = input("Guess a letter: ").lower()
        
        if len(guess) != 1 or not guess.isalpha():
            print("Enter a single letter.\n")
            continue
        
        if guess in guessed_letters:
            print("Already guessed.\n")
            continue
        
        guessed_letters.add(guess)
        
        if guess in secret_word:
            print(f"Good guess!\n")
        else:
            incorrect_guesses += 1
            print(f"Wrong.\n")
    
    print(f"Game over. The word was: {secret_word}")
```

```python
# Test functions
print("Random words:")
for i in range(3):
    print(choose_word())

print("\nTest show_word:")
print(show_word('apple', {'a', 'p', 'l'}))
```

```python
play_hangman()
```
