## Problem 1: Count Even Numbers

**Statement:** Find the total of numbers from a collection that, divided by 2, the module is 0.

```python
def count_evens(numbers):
    """
    Count how many numbers in a collection are evenly divisible by 2.
    
    Args:
        numbers: A list of numbers
        
    Returns:
        Integer count of even numbers
    """
    total = 0
    
    for num in numbers:
        if num % 2 == 0:
            total += 1
    
    return total

# Test with the provided collection
my_list = [3, 5, 27, 20, 6, 3.5, 60]
result = count_evens(my_list)
print(f"Total even numbers: {result}")
```

```python
# Additional tests
test1 = [3, 4, 10, 15, 999, 2]
print(f"Test 1: {count_evens(test1)} (expected: 3)")

test2 = [2, 4, 6, 10, 3.7]
print(f"Test 2: {count_evens(test2)} (expected: 4)")

test3 = [1, 3, 5, 7, 9]
print(f"Test 3: {count_evens(test3)} (expected: 0)")

test4 = [2, 4, 6, 8, 10]
print(f"Test 4: {count_evens(test4)} (expected: 5)")
```
